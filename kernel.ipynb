# Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Data

```python
df = pd.read_csv('../input/Attacks on Political Leaders in Pakistan.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
plt.figure(figsize=(45,20))
plt.plot(df['Date'],df['City'])
plt.ylabel('City', fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.title("Attacks on each city on specific dates", fontsize=18)
plt.show()
```

```python
provinces = df['Province'].unique()
attacks_in_provinces = []
for province in provinces:
    attacks_in_provinces.append(df[df['Province']==province].count().Province)
plt.figure(figsize=(12,8))
plt.bar(provinces,attacks_in_provinces)
plt.ylabel('Counts of Attacks', fontsize=14)
plt.xlabel('Province', fontsize=14)
plt.title("Total attacks in each province", fontsize=18)
plt.show()
```

```python
sucessful_attacks = df[df['Target Status']=='Killed']
failed_attacks = df[df['Target Status']=='Escaped']
print(f"Total Sucessful Attacks {sucessful_attacks['Target Status'].count()}")
print(f"Total Failed Attacks {failed_attacks['Target Status'].count()}")
```

```python
parties = df['Party'].unique()
attacks_on_parties = []
for party in parties:
    attacks_on_parties.append(df[df['Party']==party].count().Party)
plt.figure(figsize=(12,8))
plt.bar(attacks_on_parties,parties)
plt.ylabel('Parties', fontsize=14)
plt.xlabel('Count of Attacks', fontsize=14)
plt.title("Total attacks on each party", fontsize=18)
plt.show()
```

```python
categories = df['Target Category'].unique()
attacks_of_category = []
for category in categories:
    count_series = df[df['Target Category']==category].count()
    attacks_of_category.append(count_series['Target Category'])
plt.figure(figsize=(12,8))
plt.bar(categories, attacks_of_category)
plt.ylabel('Count of Attacks', fontsize=14)
plt.xlabel('Attack Type', fontsize=14)
plt.title("Total attack on each category", fontsize=18)
plt.show()
```

```python
total_killed = df['Killed'].sum()
total_injured = df['Injured'].sum()
totals = [total_killed, total_injured]
cats = ["Killed", "Injured"]
plt.figure(figsize=(12,8))
plt.bar(cats, totals)
plt.ylabel('Count of Killed or Injured', fontsize=14)
plt.xlabel('Killed or Injured', fontsize=14)
plt.title("Total number of killed & injured", fontsize=18)
plt.show()
```
